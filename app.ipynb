```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Descargar la p√°gina web
url = 'https://www.sii.cl/valores_y_fechas/uf/uf2023.htm'
response = requests.get(url)


```

```python
from bs4 import BeautifulSoup
import pandas as pd

soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find_all('table')[5]

# Convertir la tabla en un DataFrame de pandas
df = pd.read_html(str(table))[0]

primeros = df.iloc[:, :2]
primeros.columns = ['Dia', 'Valor']

segundos = df.iloc[:, 2:4]
segundos.columns = ['Dia', 'Valor']

terceros = df.iloc[:, 4:]
terceros.columns = ['Dia', 'Valor']

df_concat = pd.concat([primeros, segundos,terceros], axis=0)

df_concat
```

```python
def buscar_valor(Dataframe : pd.DataFrame, dia : int):
    
    return Dataframe[Dataframe['Dia'] == 25]
```

```python
buscar_valor(df_concat,1)
```

```python
def tomar_tabla(anio: int, mes: int) -> BeautifulSoup :

    '''
    El objetivo de esta funcion es hacer un scraping de la tabla.

    '''
    url = f'https://www.sii.cl/valores_y_fechas/uf/uf{anio}.htm'

    response = requests.get(url)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    tabla = soup.find_all('table')[mes]

    return tabla
```
